```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# okay, let's draw a circle

def _generate_circle(nr_data_points:int=1_000):
    vectors = np.random.randn(2, nr_data_points)
    vectors_lengths = np.linalg.norm(vectors, axis=0)
    return vectors / vectors_lengths

xs, ys = [], []
v = _generate_circle()
plt.scatter(v[0], v[1])

# let's add random noise
# with stddev = 0.1
# and equilibrium bond length = 1.0

# with stddev = 0.1
# and equilibrium bond length = 1.0
xs, ys = [], []
v = _generate_circle()

equilibrium_bond_length = 1.0
stddev = 0.1
v = v * (np.random.randn(v.shape[1]) * stddev + equilibrium_bond_length)

plt.scatter(v[0], v[1])
plt.show()
```
